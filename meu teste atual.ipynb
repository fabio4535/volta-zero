```python
# 4) seu código aqui
# Lista com as colunas de interesse
colunas_interesse = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 
                     'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']

# Selecionar apenas essas colunas
df_selecionado = sinasc[colunas_interesse]

# Remover linhas onde a coluna APGAR5 está ausente (missing)
df_apgar = df_selecionado.dropna(subset=['APGAR5'])

# Contar o número de linhas restantes após a remoção
num_linhas = len(df_apgar)

# Contar os valores ausentes nas colunas (depois da remoção de APGAR5 missing)
missing_restantes = df_apgar.isnull().sum()

# Exibir os resultados
print("Número de linhas restantes após remover APGAR5 ausente:", num_linhas)
print("\nValores missing por coluna:")
print(missing_restantes)
```

```python
# 5) seu código aqui
# Substituir o valor 9 por NaN nas colunas especificadas
colunas_para_substituir = ['ESTCIVMAE', 'CONSULTAS']
df_apgar[colunas_para_substituir] = df_apgar[colunas_para_substituir].replace(9, np.nan)

# Contar os valores ausentes (missing) após a substituição
missing_atualizado = df_apgar.isnull().sum()

# Exibir o resultado de forma clara
print("Valores ausentes atualizados (após substituir 9 por NaN):")
print(missing_atualizado)
```

```python
# 7) seu código aqui
# Selecionar colunas de interesse
colunas = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE',
           'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
df = sinasc[colunas].copy()

# Substituir códigos 'ignorados' por NaN
df.replace({'ESTCIVMAE': {9: np.nan}, 'CONSULTAS': {9: np.nan}}, inplace=True)

# Remover registros com APGAR5 ausente
df = df[df['APGAR5'].notna()]

# Estratégias de preenchimento definidas
preenchimentos = {
    'LOCNASC': 'ignorado',
    'IDADEMAE': df['IDADEMAE'].median(),
    'ESCMAE': df['ESCMAE'].mode()[0],
    'QTDFILVIVO': 0,
    'GESTACAO': df['GESTACAO'].mode()[0],
    'GRAVIDEZ': df['GRAVIDEZ'].mode()[0],
    'ESTCIVMAE': df['ESTCIVMAE'].mode()[0],
    'CONSULTAS': df['CONSULTAS'].mode()[0]
}

# Aplicar os preenchimentos
df.fillna(value=preenchimentos, inplace=True)

# Verificar se ainda há valores ausentes
print("Valores ausentes por coluna:")
print(df.isna().sum())
```

```python
# 8) seu código aqui
# Exemplo de DataFrame com a coluna Apgar
df_apgar = pd.DataFrame({
    'apgar': [9, 7, 5, 3, 10, 6, 4, 2, 8, 7, 5, 0, 9, 6]
})

# Categorização da variável Apgar
df_apgar.loc[(df_apgar['apgar'] >= 8) & (df_apgar['apgar'] <= 10), 'apgar_cat'] = 'normal'
df_apgar.loc[(df_apgar['apgar'] >= 6) & (df_apgar['apgar'] <= 7), 'apgar_cat'] = 'asfixia_leve'
df_apgar.loc[(df_apgar['apgar'] >= 4) & (df_apgar['apgar'] <= 5), 'apgar_cat'] = 'asfixia_moderada'
df_apgar.loc[(df_apgar['apgar'] >= 0) & (df_apgar['apgar'] <= 3), 'apgar_cat'] = 'asfixia_severa'

df_apgar
```

```python
# Frequência das categorias
frequencias = df_apgar['apgar_cat'].value_counts()

frequencias
```
